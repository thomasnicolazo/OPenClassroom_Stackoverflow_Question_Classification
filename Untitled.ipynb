```python
def wordFrequency(data):
    ''' take pandas series and count retun the occurences of the elements'''
    freqTotal = {}
    for text in data:
        freq = nltk.Counter(text)
        for word, count in freq.items():
            if word in freqTotal.keys():
                freqTotal[word] += count
            else:
                freqTotal[word] = count
    return freqTotal
```

```python
freqWord = wordFrequency(data['tokens_unique'])
dataFreqWord = pd.DataFrame({'word':freqWord.keys(),'count':freqWord.values()})
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
SPECIAL_TAGS = '@.'
def ListToString(liste):
    listSaveWords = []
    liste2 = liste.copy()
    for word in liste:
        if word.find('+')!=-1 or word.find('#')!=-1:
            saveWord = word
            liste2[liste.index(saveWord)] = SPECIAL_TAGS # special tags
            listSaveWords.append(saveWord)
            
    phrase = ' '.join(liste2)
    return (phrase,listSaveWords)
```

```python
#tokenWithoutSW[0:3]
```

```python
tokenWithoutSWCopy = tokenWithoutSW.copy()
TokenLemma = []
for tokenList in tokenWithoutSWCopy:
        #display(data.head())
        text, listSaveWords = ListToString(tokenList)
        #display(data.head())
        #print(text)
        #print(listSaveWords)
        doc = nlp(text)
        liste = []
        k=0
        for token in doc:
            #print(token)
            if str(token.lemma_) == SPECIAL_TAGS:
                token = listSaveWords[k]
                k=+1
                liste.append(token)
            else:
                liste.append(token.lemma_)
        TokenLemma.append(liste)

#final_string = ' '.join(map(str,empty_list))
    
```

```python
def occurenceMatrix(data, wordColumn):
    '''  search occurence of wordColumn(list) into data(pandas series) '''
    lenColumns = len(wordColumn)
    matrix = np.ndarray(shape=(lenColumns,lenColumns), dtype=np.float64())
    for indx, tag1 in enumerate(wordColumn):
        for indy , tag2 in enumerate(wordColumn):
            if tag1 != tag2:
                sumTags1=0
                sumTags2=0
                sumBinome = 0
                for j in data:
                    if tag1 in j:
                        sumTags1+=1
                        if tag2 in j:
                            sumBinome+=1
                    if tag2 in j:
                        sumTags2+=1
            else:
                sumTags1 = sumTags2 = 1
                sumBinome = 0
                
            matrix[indx,indy]=sumBinome/(sumTags1*sumTags2)
    #print(matrix)
    return matrix
                
        
```
